```python
# Mutations

def mutate_string(string, position, character):

    l = list(string)
    l[position] = character
    string = "".join(l)
    
    return string

if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```

```python
# Find a string

def count_substring(string, sub_string):
    liste = []
    n = len(sub_string)

    for i in range(0,len(string)):
        liste.append(string[i:n+i])
        
    count = liste.count(sub_string)
    return count    

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

```python
# Text Alignment

#Replace all ______ with rjust, ljust or center. 

thickness = int(input()) #This must be an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
# String Formatting

def print_formatted(number):
    
    l = len(str(bin(number))[2:])
    
    for i in range(1,number+1):
        
        print( str(i).rjust(l," "), oct(i)[2:].rjust(l," "), hex(i)[2:].upper().rjust(l, " "), bin(i)[2:].rjust(l," ") )
    
if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

```python
# Alphabet Rangoli

import string
alfabe = string.ascii_lowercase

def print_rangoli(size):
    
    kelime = alfabe[:size]
    liste = []
    
    for i in range(size):
        s = kelime[i:]
        liste.append("-".join(s[::-1]+s[1:]).center(4*size-3,"-"))
        
    print("\n".join(liste[::-1]+liste[1:]))
       
if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
# The Minion Game

def minion_game(string):
    n = len(string)
    kevin = 0
    stuart = 0
    vowel = ["A","E","I","O","U"]
    
    for i in range(n):
        if (string[i] in vowel):
            kevin += n - i
            
        else:
            stuart += n - i
            
    if (kevin > stuart):
        print("Kevin",kevin)
        
    elif (kevin < stuart):
        print("Stuart",stuart)
        
    else:
        print("Draw")

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
# Merge the Tools!

def merge_the_tools(string, k):
    
    n = len(string)
    
    for i in range(0,n,k):
        char = ""
        
        for j in string[i:i+k]:
            
            if j not in char:
                char += j
                
        print(char)
            
if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```
