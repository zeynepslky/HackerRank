```python
# Simple Array Sum


import math
import os
import random
import re
import sys

def simpleArraySum(ar):

    toplam = 0
    for i in ar:
        toplam += i
    return toplam
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    ar_count = int(input().strip())

    ar = list(map(int, input().rstrip().split()))

    result = simpleArraySum(ar)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# Compare the Triplets


import math
import os
import random
import re
import sys

def compareTriplets(a, b):

    sonuc = [0,0]
    for i in range(3):
        if (a[i] > b[i]):
            sonuc[0] += 1
        elif (a[i] < b[i]):
            sonuc[1] += 1
        else:
            continue
    
    return sonuc
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
# Plus Minus


import math
import os
import random
import re
import sys

def plusMinus(arr):

    a = 0
    b = 0
    c = 0
    d = len(arr)
    for i in arr:
        if (i>0):
            a += 1
        elif (i<0):
            b += 1
        else:
            c += 1
    
    print("{:.6f}".format(a/d))    
    print("{:.6f}".format(b/d))   
    print("{:.6f}".format(c/d))   

if __name__ == '__main__':
    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    plusMinus(arr)
```

```python
# Staircase


import math
import os
import random
import re
import sys

def staircase(n):

    for i in range(1,n+1):
        print((n-i)*" "+i*"#", sep = "\n")

if __name__ == '__main__':
    n = int(input().strip())

    staircase(n)
```

```python
# Time Conversion


import math
import os
import random
import re
import sys

def timeConversion(s):

    if (s[8]=="P" and int(s[:2]) < 12):
        return (str(int(s[:2]) + 12)+s[2:8])

    elif (s[8]=="A" and int(s[:2]) == 12):
        return "00" + s[2:8]
    else:
        return s[:8]

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```
